```python
import tkinter as tk
import tkinter.font as tkfont

def alert_message():

    print('in alert')
    def yes_action():
        print("Yes button clicked")
        global flag
        flag=1
        root.destroy()

    def no_action():
        print("No button clicked")
        root.destroy()

    # Create the main window
    root = tk.Tk()

    # Set the background color of the main window to transparent
    root.wait_visibility(root)  # Wait for window to be visible
    root.attributes("-alpha", 0.5)

    # Remove the window header
    root.overrideredirect(True)

    # Get the screen width and height
    screen_width = root.winfo_screenwidth()
    screen_height = root.winfo_screenheight()

    # Set the window size to match the screen width and height
    root.geometry("{0}x{1}+0+0".format(root.winfo_screenwidth(), root.winfo_screenheight()))

    # Create the overlay window
    overlay_window = tk.Toplevel(root)

    # Set the background color of the overlay window to light yellow
    overlay_window.configure(bg="#f9e79f")

    # Remove the window header
    overlay_window.overrideredirect(True)

    # Set the window size to 50% of the screen width and height
    window_width = int(screen_width / 1.5)
    window_height = int(screen_height / 2)
    overlay_window.geometry("{}x{}+{}+{}".format(window_width, window_height, int((screen_width - window_width) / 2), int((screen_height - window_height) / 2)))

    # Set the font for the message label
    font_style = tkfont.Font(family="Helvetica", size=36, weight="bold")

    # Create the label for the message
#     Please confirm. Are you trying to delete this {section} ???
    
    
    alert_text='Please confirm. Are you trying to \n delete '+ " {" + course_name[course_name.find(' '):] + "} "
    message_label = tk.Label(overlay_window, text=alert_text, font=font_style, fg="#e74c3c", bg="#f9e79f")
    message_label.pack(expand=True, pady=(window_height//4, 0))

    # Create an empty label
    empty_label = tk.Label(overlay_window, text="", bg="#f9e79f")
    empty_label.pack()

    # Create the frame for the buttons
    button_frame = tk.Frame(overlay_window, bg="#f9e79f")
    button_frame.pack(expand=True)

    # Set the font for the buttons
    button_font_style = tkfont.Font(family="Helvetica", size=24, weight="bold")

    # Create the Yes button
    yes_button = tk.Button(button_frame, text="Yes", font=button_font_style, bg="#f1948a", fg="#ffffff", command=yes_action)
    yes_button.pack(side=tk.LEFT, padx=5)

    # Create an empty label
    empty_label2 = tk.Label(button_frame, text="", bg="#f9e79f")
    empty_label2.pack(side=tk.LEFT, padx=5)

    # Create the No button
    no_button = tk.Button(button_frame, text="No", font=button_font_style, bg="#f1948a", fg="#ffffff", command=no_action)
    no_button.pack(side=tk.RIGHT, padx=5)
    
    print('in root')
    # Run the main loop
    root.wm_attributes("-topmost", True)
    overlay_window.wm_attributes("-topmost", True)
    root.mainloop()
    overlay_window.mainloop()
```
